### heatmap

```python
house = pd.read_csv('./seaborn_data/AmesHousing.csv')
```

```python
house.corr()
```

```python
house30 = house.iloc[:, 1:30]
```

```python
plt.figure(figsize=(15, 10)) 
heatmap = sns.heatmap(house30.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG') 

plt.yticks(size=15)
heatmap.set_title('Correlatio Heatmap', fontdict={'fontsize':18}, pad=12);
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(15, 10)) 

#  ,        True 
mask = np.triu(np.ones_like(house30.corr(), dtype=np.bool)) 
heatmap = sns.heatmap(house30.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG') 
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

```python
with open('file.txt') as f:
    f.seek(0)
    name_from = f.readline()
    name_from = name_from.rstrip('\n')
    name_to = f.readline()
    name_to = name_to.rstrip('\n')
    num = f.readline().lower()

with open(f'{name_from}') as f1:
    text = f1.readlines()
    sentence = text[int(num)]
    
with open(f'{name_to}', 'a') as f2:
    f2.write(sentence)
    
```
