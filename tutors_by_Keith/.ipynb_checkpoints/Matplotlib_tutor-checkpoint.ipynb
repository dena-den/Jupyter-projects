```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]

plt.figure(figsize=(10,6), dpi=100)

#plt.plot(x, y, label='x**x', color='#000', linewidth=3, marker='.',
#             linestyle='--', markersize=15)
plt.plot(x, y, 'b', label='x^2')

x2 = np.arange(0,6.5,0.5)
plt.plot(x2[:8], x2[:8]**3, 'r', label='x^3')
plt.plot(x2[7:], x2[7:]**3, 'r:')


plt.title('Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 22})
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.savefig('graph.png', dpi=300)

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [14, 32, 26]
plt.figure(figsize=(8,4))
bars = plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```

```python
fifa = pd.read_csv('.\matplotlib_tutorial-master\\fifa_data.csv')
```

```python
fifa.head()
```

```python
bins = np.arange(40,110,10)

plt.figure(figsize=(10,6), dpi=200)
plt.hist(fifa.Overall, bins=bins)
plt.xticks(bins)
plt.title('Player skills in FIFA')
plt.show()
```

```python
left = fifa[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels=['left', 'right'], 
       colors=['red', 'blue'], autopct='%.2f %%')
plt.title('Foot in FIFA')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) 
               if type(x) == str else x 
               for x in fifa.Weight]
```

```python
fifa.Weight
```

```python
light = fifa[fifa.Weight < 160].count()[0]
medium = fifa[(fifa.Weight >= 160) & (fifa.Weight < 180)].count()[0]
heavy = fifa[fifa.Weight >= 180].count()[0]
```

```python
plt.style.use('default')

plt.pie([light, medium, heavy])
plt.show()
```

```python
Barc = fifa[fifa.Club == 'FC Barcelona']['Overall']
Real = fifa[fifa.Club == 'Real Madrid']['Overall']
```

```python
boxes = plt.boxplot([Barc, Real], patch_artist=True,
                   medianprops={'linewidth':2})
plt.figure(figsize=(5,8))
for box in boxes['boxes']:
    box.set(color='r', linewidth=2, facecolor='y')
plt.show()
```
