```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
%matplotlib inline
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(
    df['TV'],
    df['sales'],
    c='black'
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
x = df['TV'].values.reshape(-1,1)
y = df['sales'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(x, y)
print("The linear model is: y = {:.3} + {:.3}x" \
                   .format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(x)
plt.figure(figsize=(12, 6))
plt.scatter(
    df['TV'],
    df['sales'],
    c='black'
)
plt.plot(
    df['TV'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
x_radio = df['radio'].values.reshape(-1,1)
y = df['sales'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(x_radio, y)
print("The linear model is: y = {:.3} + {:.3}x" \
                   .format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions_radio = reg.predict(x_radio)
plt.figure(figsize=(12, 6))
plt.scatter(
    df['radio'],
    df['sales'],
    c='black'
)
plt.plot(
    df['radio'],
    predictions_radio,
    c='blue',
    linewidth=2
)
plt.xlabel("Money spent on radio ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
x = df['TV']
y = df['sales']
x2 = sm.add_constant(x)
est = sm.OLS(y, x2)
est2 = est.fit()
print(est2.summary())
```

```python
xs = df.drop(['sales'], axis=1)
y = df['sales'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(xs, y)
print("The linear model is: Y = {:.3} + {:.3}*TV + {:.3}*radio + {:.3}*newspaper" \
      .format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1], reg.coef_[0][2]))
```

```python
x = np.column_stack((df['TV'], df['radio'], df['newspaper']))
y = df['sales']
x2 = sm.add_constant(x)
est = sm.OLS(y, x2)
est2 = est.fit()
print(est2.summary())
```
