#### Из жанров popular_uniq_genres будем предлагать книги

```python
popular_uniq_genres = items.genres.value_counts().to_frame()

popular_uniq_genres = popular_uniq_genres[popular_uniq_genres.genres >= 100]

pop_uniq_genres = list(popular_uniq_genres.index.unique())

len(pop_uniq_genres)
```

#### Это все уникальные жанры uniq_genres
#### Их добавляем в genres_dict по совпадениям слов

Ключи genres_dict - это жанры, в которых есть более 100 книг.
Значения genres_dict - это все жанры, у которых есть по 3 (2 или 1 в зависимости от количества слов в жанре) общих слова с основными жанрами, которые в ключах.

```python
uniq_genres = list(items.genres.unique())
```

```python
genres_dict = {} # словарь связи популярных жанров со всеми остальными

for key in pop_uniq_genres:
    genres_dict[key] = []
    key_list = key.replace(',', ' ').split(' ')
    if len(key_list) == 1:
        edge = 1
    elif len(key_list) == 2:
        edge = 2
    else:
        edge = 3
    for genre in uniq_genres:
        match = 0
        for key0 in key_list:
            if key0 in genre:
                match += 1
                if match >= edge:
                    genres_dict[key].append(genre)
                    match = -100
```

### Выбираем лучшие три книги для каждой из групп жанров

```python
best_book3_in_genres = {}
best_book2_in_genres = {}
best_book1_in_genres = {}

for key, value in genres_dict.items():
    df = items.query('genres in @value').sort_values('popularity', ascending=False)
    best_book3_in_genres[key] = df.iloc[2, 0]
    best_book2_in_genres[key] = df.iloc[1, 0]
    best_book1_in_genres[key] = df.iloc[0, 0]
    best_book1 = pd.DataFrame(list(best_book1_in_genres.items()), 
                              columns=['genre', 'best_item_id_1'])
    best_book2 = pd.DataFrame(list(best_book2_in_genres.items()), 
                              columns=['genre', 'best_item_id_2'])
    best_book3 = pd.DataFrame(list(best_book3_in_genres.items()), 
                              columns=['genre', 'best_item_id_3'])
    best_book = pd.merge(best_book1, best_book2, on='genre')
    best_book = pd.merge(best_book, best_book3, on='genre')
```

### Предсказываем.
Смотрим прочитанные книги в interactions для каждого искомого юзера. По ВСЕМ жанрам прочитанных книг смотрим, какие из "основных" жанров (те, по которым есть более 100 книг), встречаются чаще всего в прочитанных книгах пользователя. По этим самым популярным жанрам отбираем 3 самых популярных книги из всех items. И предлагаем юзеру по 3 самых популярных в книге в каждом из его любимых жанров. Если не набирается 10 книг, то оставшиеся добавляем из наиболее популярных по всем items.

```python
for user in users_to_predict.index:
    counter = pd.DataFrame(index=genres_dict.keys(), columns=['match'])
    counter.fillna(0, inplace=True)

    user_interactions = interactions[interactions.user_id == user]
    if user_interactions.empty:
        ans = [str(best.iloc[i, 0]) for i in range(0,10)]
        users_to_predict.loc[user, 'item_id'] = ' '.join(ans)
        continue
    user_interactions = pd.merge(user_interactions, items[['item_id', 'genres']],
                                 how='left', on='item_id')

    for genre in user_interactions.genres:
        for key, value in genres_dict.items():
            if genre in value:
                counter.loc[key, 'match'] += 1
                    
    counter = counter[counter.match > 0].sort_values('match', ascending=False) \
        .reset_index().rename({'index': 'genre'}, axis=1)
    counter = pd.merge(counter, best_book, on='genre', how='left')
    
    ans = []
    for i in range(0, len(counter)):
        for j in [2,3,4]:
            ans.append(str(counter.iloc[i, j]))

    temp = [] 
    for x in ans: 
        if x not in temp: 
            temp.append(x) 
    ans = temp
    
    if len(ans) > 10:
        ans = ans[:10]
    elif len(ans) < 10:
        i = 0
        while len(ans) < 10:
            add = str(best.iloc[i, 0])
            if add not in ans:
                ans.append(add)
            i += 1
    else:
        pass
    
    users_to_predict.loc[user, 'item_id'] = ' '.join(ans)
    
    print(user)
    print(users_to_predict.loc[user])
    print('\n')
```
